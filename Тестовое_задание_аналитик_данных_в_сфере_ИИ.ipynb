#Построение прогнозной модели

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from math import sqrt
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
```

Уберем лишние колонки, которые для нас неинформативны. Отметим, что данные за нужный период мы предварительно отобрали. Поскольку по условию прогнозная модель должна учитывать день недели, то необходимо учитывать данный столбец при выборе features для модели.

```python
X = merged_df.drop(columns=['qnty', 'day_sales', 'month_sales', 'year_sales', 'index', 'day_start', 'day_end', 'month_start', 'month_end', 'year_start', 'year_end', 'day_of_week_start', 'day_of_week_end'])
y = merged_df['qnty']
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train
```

Для поиска оптимальных параметров будем использовать GridSearchCV. Также в качестве альтернативы можно было использовать и Optuna.

```python
param_grid_rf = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}
```

```python
# Настройка GridSearchCV для RandomForest
rf = RandomForestRegressor(random_state=42)
grid_search_rf = GridSearchCV(estimator=rf, param_grid=param_grid_rf, cv=3, scoring='neg_mean_squared_error', n_jobs=-1)
```

```python
# Обучение и поиск лучших параметров
grid_search_rf.fit(X_train, y_train)
```

```python
# Вывод лучших параметров
print("Лучшие параметры для RandomForest:", grid_search_rf.best_params_)
```

```python
# Прогнозы и оценка модели RandomForest
best_rf = grid_search_rf.best_estimator_
y_pred_rf = best_rf.predict(X_test)
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))
print(f'RMSE для RandomForest: {rmse_rf}')
```

```python
param_grid_xgb = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 6, 10],
    'learning_rate': [0.01, 0.1, 0.3]
}
```

```python
# Настройка GridSearchCV для XGBRegressor
xgb = XGBRegressor(random_state=42, objective='reg:squarederror')
grid_search_xgb = GridSearchCV(estimator=xgb, param_grid=param_grid_xgb, cv=3, scoring='neg_mean_squared_error', n_jobs=-1)
```

```python
grid_search_xgb.fit(X_train, y_train)
```

```python
print("Лучшие параметры для XGBRegressor:", grid_search_xgb.best_params_)
```

```python
# Прогнозы и оценка модели XGBRegressor
best_xgb = grid_search_xgb.best_estimator_
y_pred_xgb = best_xgb.predict(X_test)
rmse_xgb = np.sqrt(mean_squared_error(y_test, y_pred_xgb))
print(f'RMSE для XGBRegressor: {rmse_xgb}')
```

Теперь осуществим предсказание при помощи модели XGB регрессора (он показал наилучшие результаты) на всем нашем массиве данных

```python
predictions = best_xgb.predict(X)
rmse_xgb = np.sqrt(mean_squared_error(y, predictions))
print(f'RMSE для XGBRegressor: {rmse_xgb}')
```

Присоединим наши предсказания к исходному набору features

```python
res_df = X.copy()
res_df['qnty'] = predictions
```

С агрегируем наши предсказания, сгруппировав по дате и id товара. Этого требует от нас условие задания в части представления результатов. При этом наша прогнозная модель учитывает цену товара и день недели как и требовалось
